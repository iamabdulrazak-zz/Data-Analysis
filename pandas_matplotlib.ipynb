```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("datasets/avocado.csv")
california_df = data.copy()[data["region"] == "California"]
california_df.set_index("Date", inplace=True)
california_df['AveragePrice'] = california_df['AveragePrice'].rolling(25).mean()
california_df['price25ma'] = california_df['AveragePrice'].rolling(25).mean()
california_df.plot(figsize=(15,5))
```

```python
graph_df = pd.DataFrame()
for region in data['region'].unique():
    print(region)
    region_df = data.copy()[data['region']==region]
    region_df.set_index("Date",inplace=True)
    region_df.sort_index(inplace=True)
    region_df['price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.__dict__:
        graph_df = region_df[['price25ma']]
    else:
        graph_df = graph_df.join(region_df['price25ma'])
graph_df.dropna().plot(figsize=(15,5))
```
